### Informações sobre o Dataset


O dataset Solar Flare é utilizado para classificação multioutput.

 #### Classificação:
 
 O objetivo é, a partir das 10 características preditoras (entradas), classificar em uma das três classes de erupções solares são previstas, representadas nas três últimas colunas:

 - Produção de erupções da classe C por esta região nas próximas 24 horas (erupções comuns)
 - Produção de erupções da classe M por esta região nas próximas 24 horas (erupções moderadas)
 - Produção de erupções da classe X por esta região nas próximas 24 horas (erupções severas)
 

#### Características do Dataset

- **Nome:** Solar Flare Database
- **Número de Instâncias:**

    - *flare.data1:* 323 amostras
    - *flare.data2:* 1066 amostras

- **Número de Atributos:** 13 (10 preditores + 3 variáveis-alvo)
- **Valores Faltantes:** Nenhum
- **Formato dos Atributos:**

    Categóricos: 1º, 2º e 3º atributos (classe de Zurique, tamanho da maior mancha solar e distribuição das manchas solares)
    Numéricos discretos: Os demais atributos variam de 1 a 3 (atividade, evolução, etc.)
    Variáveis-alvo: Três colunas indicam o número de flares previstos em 24 horas para classes C, M e X

- **Distribuição das Classes:**

    A maioria das instâncias tem zero flares, especialmente para M-class e X-class flares.

### Etapa 01 - Pre-processamento dos dados

## Limpeza dos Dados

#### 1. Transformação das variáveis categóricas

Vamos converter variáveis categóricas para numéricas para permitir que o modelo as utilize. Para isso, o scikit-learn possui o labelEncoder, que atribui um valor numérico para cada item categórico (text), permitindo que as operações matemáticas ocorram.

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

X['modified Zurich class'] = le.fit_transform(X['modified Zurich class'])
X['largest spot size'] = le.fit_transform(X['largest spot size'])
X['spot distribution'] = le.fit_transform(X['spot distribution'])
```

```python
X[['modified Zurich class', 'largest spot size','spot distribution']].head()
```

#### 3. Análise da Correlação dos Dados

```python
# Visualizando a correlação entre as variáveis para identificar relações fortes.

# Tabela de correlação
corr = X.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, cmap=sns.color_palette("Blues"), linewidths=.5, annot=True)
```

## Etapa 02 - Separando os dados de teste e treino

```python
# Separando os dados de treino e teste, considerando 80% para treino e 20% para teste.

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### Validação Cruzada

## Etapa 03 - Treinamento do Modelo


```python
from sklearn.multioutput import MultiOutputClassifier
from sklearn.linear_model import SGDClassifier

modelo = SGDClassifier()

# Criando o modelo MultiOutputClassifier para lidar com múltiplas saídas
classificadorSolar = MultiOutputClassifier(modelo)

# Treine o classificador multioutput
classificadorSolar.fit(x_train, y_train)
```

```python
# Fazendo predições do modelo nos dados de teste

y_pred_class = classificadorSolar.predict(x_test)
```

```python
# Avaliando o modelo

from sklearn.metrics import accuracy_score

# Acurácia para cada variável de saída
for i in range(y_test.shape[1]):
    # Acesse as colunas usando iloc ou valores
    print(f"Acurácia para a variável {i}: {accuracy_score(y_test.iloc[:, i].values, y_pred_class[:, i])*100:.2f}%")
```

Observação: A avaliação do modelo para problemas de classificação multioutput ocorre individualmente para cada classe. Isso é, no caso de y_test, temos y_test.shape[1] = 3, ou seja, vetor de comprimento 3, onde cada posição armazena, a acurácia de cada saída (classe).

```python
from sklearn.metrics import classification_report

# Gerar o relatório de classificação para cada variável de saída
for i in range(y_test.shape[1]):
    print(f"\nRelatório de Classificação para a variável {i}:\n")
    print(classification_report(y_test.iloc[:, i], y_pred_class[:, i]))

```

### Search Grid (GridSearchCV) para otimização dos parâmetros

```python
# GridSearchCV para buscar os melhores parâmetros para o nosso modelo de classificação.

# Definindo o grid de parâmetros para o SGDClassifier
param_grid = {
    'estimator__loss': ['hinge', 'log', 'modified_huber'],
    'estimator__penalty': ['l2', 'l1', 'elasticnet'],
    'estimator__alpha': [0.0001, 0.001, 0.01, 0.1],
    'estimator__learning_rate': ['constant', 'optimal', 'invscaling'],
    'estimator__max_iter': [1000, 2000, 3000]
}
```

```python
# Criando o objeto GridSearchCV
grid_search = GridSearchCV(estimator=classificadorSolar,
                           param_grid=param_grid, cv=5, n_jobs=-1, verbose=1)

# Ajustar o GridSearchCV aos dados de treinamento
grid_search.fit(x_train, y_train)
```

```python
# Exibindo os melhores parâmetros encontrados
best_model = grid_search.best_estimator_
print(f"Melhores parâmetros: {grid_search.best_params_}")
```

```python
# Obter o melhor modelo treinado
best_clf = grid_search.best_estimator_
```

```python
# Acurácia por saída (coluna)
for i in range(y_test.shape[1]):
    print(f"Acurácia para a saída {i}: {accuracy_score(y_test.iloc[:, i], y_pred_class[:, i])*100:.2f}%")
```

# Teste Prático do Modelo

```python
# Gerando dados aleatórios para os preditores (10 observações como exemplo)
np.random.seed(42)

# Simulação de dados inseridos pelo usuário (Fonte: Autores)

dados_entrada = pd.DataFrame({
    'modified Zurich class': ['H', 'F', 'H', 'D', 'D'],
    'largest spot size': ['S', 'K', 'H', 'S', 'A'],
    'spot distribution': ['O', 'X', 'C', 'I', 'O'],
    'activity': [1, 2, 1, 2, 1],
    'evolution': [1, 2, 3, 3, 3],
    'previous 24 hour flare activity': [1, 2, 3, 3, 1],
    'historically-complex': [2, 1, 2, 1, 1],
    'became complex on this pass': [1, 2, 1, 2, 2],
    'area': [1, 2, 1, 2, 1],
    'area of largest spot': [2, 2, 5, 2, 6]
})


# Codificando as variáveis categóricas com LabelEncoder (novamente há o uso do label encoder pois existema dados categóricos

le = LabelEncoder()
dados_entrada['modified Zurich class'] = le.fit_transform(dados_entrada['modified Zurich class'])
dados_entrada['largest spot size'] = le.fit_transform(dados_entrada['largest spot size'])
dados_entrada['spot distribution'] = le.fit_transform(dados_entrada['spot distribution'])

# Prevendo as classes com o modelo já treinado
y_pred = classificadorSolar.predict(dados_entrada)

# Convertendo as previsões para um DataFrame (apenas para a saida ficar padronizada)
predicao_classe = pd.DataFrame(y_pred, columns=['Classe_C', 'Classe_M', 'Classe_X'])
```

```python
print("Previsão das classes de erupções solares para os dados inseridos:")
print(predicao_classe)

# Convertendo as previsões para índices numéricos e identificando a classe com maior probabilidade
classe_final = np.argmax(y_pred, axis=1)

# Mapeando as classes numéricas de volta para suas representações textuais
classes_mapeadas = ['C', 'M', 'X']
classe_final_texto = [classes_mapeadas[i] for i in classe_final]
```

O modelo fez uma previsão em termos de 3 classes: Classe_C, Classe_M, e Classe_X, onde cada coluna representa uma dessas classes. As previsões 0 e 1 indicam a presença ou não de cada classe para cada observação. 
Ou seja:
- Classe_C é representada pela primeira coluna.
- Classe_M pela segunda coluna.
- Classe_X pela terceira coluna.

Para a primeira linha (0 0 0), significa que nenhuma das classes foi prevista para a observação correspondente (ou seja, nenhuma erupção de classe C, M ou X foi prevista para a primeira observação).
